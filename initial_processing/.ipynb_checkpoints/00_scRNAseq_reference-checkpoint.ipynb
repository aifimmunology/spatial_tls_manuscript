## Download files from HISE

```python
def cache_and_unzip_from_HISE(file_ids, data_input_folder):
    
    # cache files to input folder
    filepaths = hp.cache_files(file_ids)

    # move files to data_input_folder
    for file in filepaths:
        if os.path.exists(file):  
            shutil.copy2(file, data_input_folder)
            print(f"Copied: {file} -> {data_input_folder}")

            # Unzip the file
            file_name = os.path.basename(file)  # Extract filename from path
            zip_path = os.path.join(data_input_folder, file_name)  # Full path to copied file
            
            if zipfile.is_zipfile(zip_path):  # Check if it's a valid zip file
                with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                    extract_path = data_input_folder
                    zip_ref.extractall(extract_path)
                    print(f"Unzipped: {zip_path} -> {extract_path}")

                # Delete the .zip file after extraction
                os.remove(zip_path)
                print(f"Deleted: {zip_path}")
            else:
                print(f"Not a valid zip file: {zip_path}")

        else:
            print(f"File not found: {file}")
```

```python
def h5_to_adata(dir, sample_ID):

    # file names    
    h5_file = os.path.join(dir, 'cell_feature_matrix.h5')
    cells_file = os.path.join(dir, 'cells.csv.gz')
    
    # h5 to adata
    adata = sc.read_10x_h5(
        filename=h5_file
    )
    
    # cells file
    df = pd.read_csv(
        cells_file, 
        compression = 'gzip'
    )
    
    # set indx
    df.set_index(adata.obs_names, inplace=True)
    adata.obs = df.copy()
    
    # x, y
    adata.obsm["spatial"] = adata.obs[["x_centroid", "y_centroid"]].copy().to_numpy()

    # add sample
    adata.obs['sample'] = sample_ID
    
    return adata
```

### Create reference adata object

```python
adata_ref = ad.read_csv(umi_filepath)
adata_ref = adata_ref.transpose()
cell_metadata_df = pd.read_csv(
    cells_filepath, 
    compression = 'gzip', index_col=0
)
adata_ref.obs_names.equals(cell_metadata_df.index)
adata_ref.obs = cell_metadata_df
```

## Process reference adata

### First subset to xenium genes

```python
# subset to xenium genes
sample_genes_bool = adata_ref.var.index.isin(adata_sample.var.index)
adata_ref_xen = adata_ref[:, sample_genes_bool]
```

```python
print(adata_ref_xen.shape)
```

### Next, normalize, log, and reduce dimensions

```python
def scanpy_process(adata):
    # skipping qc and filtering because already done for this reference

    # norm and log
    adata.layers["counts"] = adata.X.copy()
    sc.pp.normalize_total(adata, target_sum=1e6)
    sc.pp.log1p(adata)
    
    ## dimensionality reduction 
    sc.pp.pca(adata)
    print('starting neighbors...')
    sc.pp.neighbors(adata)
    print('starting umap...')
    sc.tl.umap(adata)
    # leiden
    print('starting leiden res1')
    sc.tl.leiden(adata, resolution = 1, key_added='leiden_res1', n_iterations=2)
    return adata
```

```python
adata_ref_xen = scanpy_process(adata_ref_xen)
```
