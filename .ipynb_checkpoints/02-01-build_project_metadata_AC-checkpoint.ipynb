### 1.2 Custom Functions
> It ain't much but its honest work. 

````python
def add_time_stamp(df):
    fname = df['file.name']
    fname_split = [f.split('/') for f in fname]
    ts = [x[2] for x in fname_split]
    df['manual.time_stamp'] = ts
    return df
````

````python
def find_file_paths_with_extension(directory, extension):
    matches = []
    for root, dirnames, filenames in os.walk(directory):
        for filename in fnmatch.filter(filenames, '*.' + extension):
            matches.append(os.path.join(root, filename))
    return matches
````

````python
def find_file_path_with_name(start_dir, target_file):
    for root, _, files in os.walk(start_dir):
        if target_file in files:
            return os.path.join(root, target_file)
    return None
````

> Here, I'm picking out the columns I need for the scRNAseq data. Any column I add to the object, I prepend with the term 'manual' to let me know its not autogenerated by HISE scRNAseq. 

````python
fields = [
    "cmv.ab_screen_index_value",
    "cmv.ab_screen_result",
    "cmv.ab_visit",
    "igg.iga",
    "igg.igg",
    "igg.igm",
    "igg.immunofixation",
    "igg.k_flc",
    "igg.kl_flc_ratio",
    "igg.l_flc",
    "igg.mono_1_id",
    "igg.mono_1_quant",
    "lip.cholesterol_hdl_ratio",
    "lip.cholesterol_hdl",
    "lip.cholesterol_ldl",
    "lip.cholesterol_non_hdl",
    "lip.cholesterol_total",
    "lip.triglycerides",
    "RISS.stage",
    "cyto.17p_loss",
    "cyto.1q_gain",
    "cyto.1p_loss",
    "cyto.high_risk",
    "cyto.karyotype",
    "cyto.t11_14",
    "cyto.t14_16",
    "cyto.t14_20",
    "cyto.t4_14",
    "cyto.karyotype"
]

subset_labs = manual_metadata[[   
    "Subject",
    "sampleKitGuid", 
    "visitDetails",
    *fields
]].fillna("Not Recorded")
````

````python
subset_labs
````

> I'll melt and map the response data to the scRNAseq metadata from the manually curated dataset. This way, I'm handling both the visit date, response type and the response of each subject at each time point. 

````python
subset_response = manual_metadata[['Subject',
                    'sampleKitGuid', 
                    'visitDetails',
                    'tb.end_induction_response', 
                    'tb.post_trans_90d_response', 
                    'tb.PostTX_1Yr_response', 
                    'tb.PostTX_2Yr_response'
                                   ]].fillna("Not Recorded")

df_melted = subset_response.melt(id_vars=['Subject', 'sampleKitGuid', 'visitDetails'], var_name='ResponseType', value_name='Response')
df = df_melted[df_melted['Response'] != 'Not Recorded']
df.reset_index(drop=True, inplace=True)
df
````

> I'll use the specimenGuid to sort BMMC and PBMC samples. This is a vulnerable process that's specific to my dataset and the files I have. We'll also be dropping some of the files in this study that we dont need. 

````python
fh1_samples = add_time_stamp(fh1_samples)
fh1_samples = fh1_samples[fh1_samples['sample.sampleKitGuid'] != 'KT01500']
fh1_samples = fh1_samples[~fh1_samples['sample.visitDetails'].isin([
    'N/A - stand-alone collection', 
    'Other'])]

fh1_samples['manual.tissue'] = fh1_samples['specimen.specimenGuid'].apply(lambda x: 
    'PBMC' if any(s in x for s in ['PB', 'PL', 'CT', 'BL']) else
    'BMMC' if any(s in x for s in ['BM', 'SP']) else
    str(x)
)

if any('MM' in str(value) for value in fh1_samples['sample.visitDetails']):
    fh1_samples['manual.disease_condition'] = 'multiple myeloma'

remove_cols = ['file.fileType', 
               'file.majorVersion', 'file.userTags.details', 'file.userTags.group', 
               'file.userTags.name','file.userTags.origin', 'file.userTags.other', 
               'file.userTags.version']

fh1_samples = fh1_samples.drop(remove_cols, axis=1)
````

````python
mapping_dicts = {
    'manual.response': dict(zip(df['sampleKitGuid'], df['Response'])),
    'manual.response_type': dict(zip(df['sampleKitGuid'], df['ResponseType']))
}

# Apply the mappings
for new_col, mapping_dict in mapping_dicts.items():
    fh1_samples[new_col] = fh1_samples['sample.sampleKitGuid'].map(mapping_dict)
````

````python
fh1_samples
````

> Samples were run more than once so there was an issue with duplicates. I'll keep the rows that have the latest date. 

````python
fh1_samples['manual.time_stamp'] = pd.to_datetime(fh1_samples['manual.time_stamp']).dt.date
dupe_check = fh1_samples.groupby(['specimen.specimenGuid', 'manual.tissue', 'sample.drawDate']).size()
dupe_check = dupe_check.reset_index(name='counts')
````

````python
dupe_check[dupe_check['counts'] == 2]
````

````python
fh1_samples = fh1_samples.sort_values(by=['specimen.specimenGuid', 'manual.time_stamp'],  ascending=[True, False])
fh1_samples = fh1_samples.drop_duplicates(subset=['specimen.specimenGuid'], keep='last')
fh1_samples.head()
````

````python
dupe_check = fh1_samples.groupby(['specimen.specimenGuid', 'manual.tissue', 'sample.drawDate']).size()
dupe_check = dupe_check.reset_index(name='counts')
dupe_check[dupe_check['counts'] == 2]
````

> Pulling out the `pbmc_sample_id` field to match that of the HISE scRNA metadata file format. This is not intuitive on the first run but once you've checked out the files it makes more sense. 

````python
# Convert float values to strings before applying regex
fh1_samples['pbmc_sample_id'] = [re.search('(PB|BMC|SP)[0-9]+-[0-9]+', file_name).group() if re.search('(PB|BMC|SP)[0-9]+-[0-9]+', file_name) is not None else None for file_name in fh1_samples['manual.file_paths']]
````

> Here's that 'manual' prepend I mentioned before. Let's add this in before merging the lab data to the subject metadata. 

````python
subset_labs_prefixed = subset_labs.add_prefix("manual.")
````

````python
merged_table = pd.merge(
    fh1_samples,
    subset_labs_prefixed,
    left_on = "sample.sampleKitGuid",
    right_on = "manual.sampleKitGuid",
    how = "left"  # Use 'outer' to keep all values from both tables
)
````

````python
merged_table.columns
````

````python
merged_table = merged_table.drop(['manual.Subject','manual.sampleKitGuid', 'manual.visitDetails'], axis=1)
````

## 5 Some metadata visualizations

````python
file_name = 'FH1_Assembled_Metadata_Manually_Cleaned_Jan_2025.csv'
file_path = find_file_path_with_name('../../input/', file_name)
data = pd.read_csv(file_path)
````

````python
data = data[~data['visitDetails'].isin(['N/A - Flu-Series Timepoint Only', 'N/A - stand-alone collection', 'Other'])]

visit_order = [
    'MM Pre-Treatment', 
    'MM Post Induction 2-Cycles', 
    'MM End Induction 1st Draw', 
    'MM End Induction 2nd Draw',
    'MM Post Transplant 60 Days', 
    'MM Post Transplant 90 Days', 
    'MM Post Transplant 1 year', 
    'MM Post Transplant 2 year'
]

data['visitDetails'] = pd.Categorical(data['visitDetails'], categories=visit_order, ordered=True)

visit_details_counts = data.groupby(['Subject', 'visitDetails']).size().unstack(fill_value=0)
visit_details_counts = visit_details_counts[visit_order]

ax = visit_details_counts.plot(kind='bar', stacked=True, figsize=(18, 6), colormap='tab20')
for container in ax.containers:
    labels = [int(v) if v > 0 else '' for v in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='center')

plt.xlabel('Subject')
plt.ylabel('Number of Visits')
plt.title('Subject Stacked by Visit Details')
plt.xticks(rotation=90)
plt.legend(title='Response', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()

plt.show()
````

````python
subset_response = data[['Subject',
                    'sampleKitGuid', 
                    'visitDetails',
                    'tb.end_induction_response', 
                    'tb.post_trans_90d_response', 
                    'tb.PostTX_1Yr_response', 
                    'tb.PostTX_2Yr_response'
                                   ]].fillna("Not Recorded")

df_melted = subset_response.melt(id_vars=['Subject', 'sampleKitGuid', 'visitDetails'], var_name='ResponseType', value_name='Response')
df = df_melted[df_melted['Response'] != 'Not Recorded']
df.reset_index(drop=True, inplace=True)
df

# Define the order of visitDetails
visit_order = ['MM End Induction 1st Draw', 'MM Post Transplant 90 Days', 'MM Post Transplant 1 year', 'MM Post Transplant 2 year']

# Convert visitDetails to a categorical type with the specified order
df['visitDetails'] = pd.Categorical(df['visitDetails'], categories=visit_order, ordered=True)

# Sort DataFrame by visitDetails and then by Subject
df = df.sort_values(by=['visitDetails', 'Subject'])

# Create scatter plot
plt.figure(figsize=(10, 5), dpi=500)
sns.scatterplot(data=df, x='Subject', y='visitDetails', hue='Response', style='Response', s=200)

# Customize the plot
plt.xticks(rotation=90)
plt.ylabel('Visit Details')
plt.xlabel('Subject')
plt.tight_layout()
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7, zorder=0)
plt.legend(title='Response', bbox_to_anchor=(1, 1), loc='upper left')

# Show the plot
plt.show()
````
